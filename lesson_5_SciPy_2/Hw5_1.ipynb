```python
from scipy.optimize import minimize
```

```python
def fence_length(arg):
    x, y = arg
    return 2 * x + y
```

```python
def constraint(arg):
    x, y = arg
    return x * y - 1000
```

```python
xbounds = (1, 1000)
ybounds = (1, 1000)
bounds = (xbounds, ybounds)
```

```python
constraints = {'type': 'ineq', 'fun': constraint}
```

```python
minimize(fence_length, [10, 10], bounds=bounds, constraints=constraints)
```
