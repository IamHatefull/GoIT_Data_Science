```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = {
    'moment' : [1, 2, 3, 4, 5, 6],
    'Yield_A' : [25, -10, 10, 5, 35, 13],
    'Yield_B' : [0, 15, -5, 5, 20, 25],
    'Yield_C' : [10, 25, -15, -5, -5, 15]
}

data = pd.DataFrame(df)
```

```python
sns.lmplot(x='Yield_A', y='Yield_B', data = data)
```

```python
sns.lmplot(x='Yield_B', y='Yield_C', data = data)
```

```python
sns.lmplot(x='Yield_C', y='Yield_A', data = data)
```
