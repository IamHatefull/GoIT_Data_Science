```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x0 = (0,0)
x1 = (10, 0)
x2 = (5, np.sqrt(75))
z = (3,4)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(111)
ax1.set_title('triangle')
plt.xlabel('X')
plt.ylabel('Y')

points = [x0, x1, x2]

X = np.array([x0[0], x1[0], x2[0]])
Y = np.array([x0[1], x1[1], x2[1]])

while True:

  index = np.random.choice(3)
  z  = ((z[0]+points[index][0])/2, (z[1]+points[index][1])/2)
  X= np.append(X,z[0])
  Y= np.append(Y,z[1])
  if len(X) == 1000:
    break

plt.scatter(X,Y)
plt.draw()
```
